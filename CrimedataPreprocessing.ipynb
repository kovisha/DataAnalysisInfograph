Reshaping the dataset for analysis

```python
# Melt the DataFrame to reshape it into long format
melted_df = pd.melt(combined_df, id_vars=['Crime Category', 'Total', 'Year'], var_name='Area', value_name='Crime Count')

# Sort the melted DataFrame by 'Crime Count' in descending order to identify the highest impact area
highest_impact_area = melted_df.groupby('Area')['Crime Count'].sum().idxmax()

print("Highest Impact Area of Crimes:", highest_impact_area)
```
